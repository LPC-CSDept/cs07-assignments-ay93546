```python
coordinate = input("Enter coordinates like 'x, y':")
x, y = coordinate.split(',')
x = int(x)
y = int(y)
if x > 0:
  if y > 0:
    print("Quadrant 1")
  else:
    print("Quadrant 4")
else:
  if y > 0:
    print("Quadrant 2")
  else:
    print("Quadrant 3")
```

```python
import random
num1 = random.randint(0, 100)
num2 = random.randint(0, 100)
num3 = random.randint(0, 100)

print(num1, num2, num3)
if num1 == num2:
    if num2 == num3:
        print("all values are the same")
    else:
        print("same:", num1, num2)
        print("different:", num3)
elif num1 == num3:
    print("same:", num1, num3)
    print("different:", num2)
elif num2 == num3:
    print("same:", num2, num3)
    print("different", num1)
else:
    print("all values are distinct")
    
```

```python
import random
a = random.randint(0, 100)
b = random.randint(0, 100)
c = random.randint(0, 100)
if a < b:
  if a < c:
    lowest_number = a
  else:
    lowest_number = c
else:
  if b < c:
    lowest_number = b
  else:
    lowest_number = c

print(a, b, c)
print(lowest_number)
```
