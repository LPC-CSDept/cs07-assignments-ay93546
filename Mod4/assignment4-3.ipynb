```python
user_input = input("Enter character: ")
while user_input != 'q':
  print(user_input, end=' ')
  user_input = input("Enter character: ")
print("loop ended because of q")
```

```python
import random
flag = 1
prev = random.randint(0,100)
while flag:
  next = random.randint(0,100)
  print("Prev:", prev, "Next:", next)  
  if next > prev:
    flag = 0
  prev = next
```
