```python
import random
prev = random.randint(0,100)
while True:
  next = random.randint(0,100)
  print("Prev:", prev, "Next:", next)  
  if next > prev:
    break
  prev = next
```

```python
import random
sum = 0
count = 0
gen_nums = []
while count < 10:
  gen_num = random.randint(0, 100)
  sum += gen_num
  gen_nums.append(gen_num)
  count += 1
  if sum > 100:
    gen_nums.pop()
    print("Numbers:", gen_nums)
    print("Total:", sum-gen_num)
    break
else:
  print("The total is <= 100")
```
