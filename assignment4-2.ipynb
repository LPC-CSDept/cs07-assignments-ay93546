```python
import random
while True:
  gen_num = random.randint(0, 100)
  if gen_num % 2 == 0:
    break
  else:
    print(gen_num)
```

```python
import random
sum = 0
while sum < 100:
  gen_num = random.randint(0,100)
  sum += gen_num
  if sum < 100:
    print(gen_num, end=' ')
print()
print("Total:", sum-gen_num)
```
