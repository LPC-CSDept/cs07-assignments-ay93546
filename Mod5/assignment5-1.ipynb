```python
# 15 20 0 3 -1 (input)
num = []
total = 0
while True:
  user_input = int(input("Enter int: "))
  if user_input < 0:
    break
  num.append(user_input)
  total += user_input
if len(num) > 2:
  total = total - min(num) - max(num)
  average = total / (len(num)-2)
else:
  total = 0
  average = 0.00
print("Sum:", total, "\t", "Average: {:.2f}".format(average))
```

```python
# 7 10 5 20 0 40 45 50 (input)
N = int(input("Enter number of list elements: "))
list = []
for i in range(N):
    list.append(int(input("Enter list values: ")))

average = sum(list)/N
for number in list:
    print("{:.2f}".format(abs(number-average)))
```

```python
#7 10 5 20 0 40 45 50 3 40 10 5
N = int(input("Enter number of list 1 elements: "))
num1 = []
for i in range(N):
    num1.append(int(input("Enter list values: ")))
M = int(input("Enter number of list 2 elements: "))
num2 = []
for i in range(M):
    num2.append(int(input("Enter list values: ")))
flag = 0
for i in range(len(num2)):
  if num2[i] in num1:
    continue
  else:
    flag = 1
    break
if flag:
  print("False")
else:
  print("True")
```
